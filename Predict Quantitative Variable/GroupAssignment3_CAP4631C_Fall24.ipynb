```python
predictor_list_2= abalone_df_without_sex.columns.difference(['Shell_weight','Rings'])
predictor_list_2
```

```python
X_all.info()
```

```python
# Polynomial features for 'horsepower'
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly_predictor = poly.fit_transform(X_all[['Shell_weight']])

# List to store adjusted R-squared values for each combination of poly-horsepower and predictor
result_list = []

# Loop through each predictor
for loop_predictor in predictor_list_2:

    # Add the current predictor to the feature matrix
    X_poly_plus_x = np.column_stack((X_poly_predictor , X_all[[loop_predictor]]))

    # Fit a linear regression model
    model = LinearRegression()
    model.fit(X_poly_plus_x, y)

    # Calculate adjusted R-squared
    y_pred = model.predict(X_poly_plus_x)
    r2_value_loop= r2_score (y, y_pred)
    adj_r_sq_model= np.round(1 - (1- r2_value_loop) * ( (y.size - 1) / (y.size - 3 - 1) ), 4) # 3 bc we have 3 predictors: hpw, hpw sq, and the iteration predictor

    # Append the results to the list
    result_list.append({"Predictor_added_to_poly": loop_predictor, "Adj_R_Squared": adj_r_sq_model})

# Create a data frame from the list of dictionaries
adding_to_poly_df = pd.concat([pd.DataFrame(item, index=[0]) for item in result_list], ignore_index=True)

adding_to_poly_df =adding_to_poly_df.sort_values(by='Adj_R_Squared', ascending=False)
```

# Justification

##The highest Adjusted R-squared is obtained for the model that combines the second-degree polynomial of Shell_weight with the linear term of Shucked_weight. Adding the linear term of Shucked_weight to the polynomial model based on Shell_weight produces a significant increase in Adjusted R-squared. Adjusted R-squared increases from 40.8% to 51.2% (a 25.46% increase). This substantial improvement justifies the inclusion of Shucked_weight in the model, as it contributes meaningfully to the model’s explanatory power.
